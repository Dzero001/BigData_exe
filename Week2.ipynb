```python
list = [[1, 'Aung', 500, 12],
        [2, 'Chip', 400, 13],
        [3, 'Min', 300 , 11]]
df = pd.DataFrame(list, columns = ['id', 'name','pay','hours'])
df
```

```python
df['income']= df['pay']*df['hours']
df
```

```python
df.drop('income', axis=1, inplace=True)
df
```

```python
df[df['pay'] >= 400]
```

```python
df[(df['pay'] >= 400) & (df['hours']< 13)]
```
